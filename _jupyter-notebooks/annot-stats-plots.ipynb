```python
df = pd.read_excel(r'E:\annot_stats\manual_monthly.xlsx')

f, axes = plt.subplots(4, 2, sharex=True, figsize=(13, 10))

CB_df = df[df['site_name_comb'] == 'CB']
KK_df = df[df['site_name_comb'] == 'KK']
ULU_df = df[df['site_name_comb'] == 'ULU17-19']
ULU22_df = df[df['site_name_comb'] == 'ULU22']
ULU23_df = df[df['site_name_comb'] == 'ULU23']
PP_df = df[df['site_name_comb'] == 'PP']

hue_order = ['B', 'BY']
palette = ["#4269D0", "#FF725C"]
sns.set_palette(palette)

palette2 ={"KK": "#EFB118", "ULU": "#4269D0", "ULU22": "#3CA951", "CB": "#97BBF5", "PP": "#9498A0", 'ULU23': '#A463F2'}

sns.histplot(ax=axes[0, 0], x=CB_df['month'], hue=CB_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[0, 0].title.set_text('Cape Bathurst')
ax0 = axes[0,0].twinx()
sns.histplot(ax=ax0, x=CB_df['month'], hue=CB_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')
ax0.set_ylabel('')

sns.histplot(ax=axes[0, 1], x=KK_df['month'], hue=KK_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, edgecolor=None)
axes[0, 1].legend(['BY', 'B'], title='Legend', bbox_to_anchor=(1.3, 1.05))
axes[0, 1].title.set_text('Kitikmeot')
ax1 = axes[0,1].twinx()
sns.histplot(ax=ax1, x=KK_df['month'], hue=KK_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')

sns.histplot(ax=axes[1, 0], x=ULU_df['month'], hue=ULU_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[1, 0].title.set_text('Ulukhaktok 2017-2019')
ax2 = axes[1,0].twinx()
sns.histplot(ax=ax2, x=ULU_df['month'], hue=ULU_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')
ax2.set_ylabel('')

sns.histplot(ax=axes[1, 1], x=ULU22_df['month'], hue=ULU22_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[1, 1].title.set_text('Ulukhaktok 2022')
ax3 = axes[1,1].twinx()
sns.histplot(ax=ax3, x=ULU22_df['month'], hue=ULU22_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')

sns.histplot(ax=axes[2, 0], x=ULU23_df['month'], hue=ULU23_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[2, 0].title.set_text('Ulukhaktok 2023')
ax4 = axes[2,0].twinx()
sns.histplot(ax=ax4, x=ULU23_df['month'], hue=ULU23_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')
ax4.set_ylabel('')

sns.histplot(ax=axes[2, 1], x=PP_df['month'], hue=PP_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[2, 1].title.set_text('Pearce Point')
ax5 = axes[2,1].twinx()
sns.histplot(ax=ax5, x=PP_df['month'], hue=PP_df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')

sns.histplot(ax=axes[3, 0], x=df['month'], hue=df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None)
axes[3, 0].title.set_text('All')

my_xticks = ['b', 'Feb', 'Apr', 'Jun', 'Aug', 'Oct', 'Dec']
axes[3,0].set_xticklabels(my_xticks)
axes[3,0].set(xlim=(0.5, 13))
axes[3, 0].set_xlabel('Month')

ax2 = axes[3,0].twinx()
sns.histplot(ax=ax2, x=df['month'], hue=df['call_type'], bins=range(0,13), hue_order=hue_order, alpha=1, multiple="dodge", discrete=True, legend=False, edgecolor=None, stat='percent')

plt.setp(axes[0, 0].get_xticklabels(), visible=False)
plt.setp(axes[0, 1].get_xticklabels(), visible=False)
plt.setp(axes[1, 0].get_xticklabels(), visible=False)
plt.setp(axes[1, 1].get_xticklabels(), visible=False)
axes[0, 0].set_xlabel('')
axes[1, 1].set_ylabel('')
axes[0, 1].set_ylabel('')
axes[2, 1].set_ylabel('')
axes[0, 0].set_ylabel('# of Annotations')
axes[1, 0].set_ylabel('# of Annotations')
axes[2, 0].set_ylabel('# of Annotations')
axes[3, 0].set_ylabel('# of Annotations')

f.delaxes(axes[3,1])
axes[2, 1].tick_params(labelbottom=True)

plt.savefig(r'E:\annot_stats\all-annots.png', bbox_inches='tight', pad_inches=0)
```
